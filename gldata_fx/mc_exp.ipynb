```python
db = dbs['dt'].get('db_0', {})
iplb = []
ind = 0  # 设置初始值为0
while True:
    tb = f"table_{ind}"
    if tb in db:
        for fd in db[tb]:
            if fd[:6] == 'field_':
                iplb.append(db[tb][fd]['name'])
        ind += 1
    else:
        break

tb8 = iplb
len(iplb)
```

```python
def is_any_chinese(strs):
    for _char in strs:
        if '\u4e00' <= _char <= '\u9fa5':
            return True
    return False
def mcfy(lb, js):
    res = ''
    for i in lb:
        if i and i.lower() in js:
            res += js[i.lower()]+'__'
        elif i.isdigit():
            res += i+'__'
        elif len(i) < 3:
            res += i+'__'
        elif is_any_chinese(i):
            res += i+'__'
        else:
            print('目前该词汇字典没有收纳：'+i)
    return res
res = mcfy(['lbs','10'], cihsy)
res
```
