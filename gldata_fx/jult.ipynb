```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def plt_kerr(x,m,n):
    # 假设你已经有了X数据
    SSE = []  # 存放每个k值下的SSE
    for k in range(m, n):
        estimator = KMeans(n_clusters=k)  # 构造聚类器
        estimator.fit(x)
        SSE.append(estimator.inertia_)  # estimator.inertia_是计算SSE  # 计算SSE

    # 绘制SSE - k图像
    plt.figure(figsize=(6, 3))
    plt.plot(range(m, n), SSE, marker='o')
    plt.xlabel("K")
    plt.ylabel("SSE")
    plt.show()
plt_kerr(x,250,270)
```
