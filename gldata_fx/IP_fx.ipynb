```python
def split_english(text):
    # 初始化结果列表
    result = [text[0]]
    bj = False
    # 遍历字符串，将大写字母作为分隔符进行分词
    for i in range(1,len(text)):
        if bj:
            result.append("_") 
            bj=False
        elif text[i].isupper():
            if i != len(text)-1:
                if not(text[i-1].isupper()) or text[i+1].islower():
                    result.append("_")    
            elif not(text[i-1].isupper()) :
                result.append("_")
        elif text[i].isdigit():
            if not text[i-1].isdigit():
                result.append("_")   
            if i != len(text)-1 and not text[i+1].isdigit():
                bj = True
        result.append(text[i])
    # 将结果列表转换为字符串并返回
    return "".join(result).split('_')

# split_english('YBICDia_gnosisYB345')
split_english('Orders0702bak')
```
